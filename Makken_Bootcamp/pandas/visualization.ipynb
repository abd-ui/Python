```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
x = np.linspace(1,20,100)
y = np.exp(x)

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, 'r*')
plt.show()
```

```python
x = data['total_bill']
y = data['tip']

plt.scatter(x,y)
plt.xlabel('Total bill$')
plt.ylabel('Tips')
plt.show()
```

```python
plt.scatter(x, y, c=y/x)
plt.xlabel('Total bill$')
plt.ylabel('Tips')
plt.show()
```

**The plot shows the relationship between total bill and tips; the brighter the color the bigger the tip percentege**

```python
sns.scatterplot(data=data, x='total_bill', y='tip', hue='day')
plt.show()
```

```python
plt.figure(figsize=(60,60))
plt.subplot(7,7,1)
sns.scatterplot(data=data, x='total_bill', y='tip', hue='day')
plt.subplot(7,7,2)
sns.scatterplot(data=data, x='total_bill', y='tip', hue='sex')

plt.show()
```

```python
sns.stripplot(data = data , x='time',y='tip',hue='day')
```

```python
male_data = data[data['sex'] == 'Male']
female_data = data[data['sex'] == 'Female']

plt.figure(figsize=(60,60))
plt.subplot(7,7,1)
sns.scatterplot(data=male_data, x='total_bill', y='tip')
plt.title('Male')
plt.subplot(7,7,2)
sns.scatterplot(data=female_data, x='total_bill', y='tip',c='r')
plt.title('Female')

plt.show()
```

```python
plt.hist(x=(data['tip']/data['total_bill'])*100)
plt.xlabel('Tip percentage %')
plt.show()
```

**Most people tip between 10 and 20 percent**

```python
plt.hist(data['total_bill'])
plt.xlabel('Total bill amount $')
plt.show()
```

```python
import plotly.express as px

fig = px.scatter_3d(data, x='total_bill', y='tip', z=data['tip']/data['total_bill'],)
fig.show()
```

```python
fig = px.scatter(data, x='total_bill', y='tip', color='day')

fig.show()
```

```python
fig = px.box(data, x='day', y='total_bill', color='time')

fig.show()
```

```python
fig = px.histogram(data, x='total_bill', y='tip', color='sex', facet_row='time',
                   facet_col='day')

fig.show()
```
