```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python

#This dataset provides detailed information about used cars, 
#including their specifications and pricing details. The dataset consists of approximately 8015 rows and 9 columns

df = pd.read_csv("cars_24_combined.csv")
df.head()
```

```python
df.dtypes
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
df.describe()
```

```python
# remove outliers
# Calculate the IQR
q1 = df["Distance"].quantile(0.25)
q3 = df["Distance"].quantile(0.75)
iqr = q3 - q1

# Define the lower and upper bounds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Remove the outliers
df_new = df.loc[(df["Distance"] >= lower_bound) &
                (df["Distance"] <= upper_bound)]
```

```python
fig = px.scatter(df_new, x="Distance", y="Price",
                 color="Year", hover_name="Car Name")

fig.show()
```

```python
fig = px.scatter(df_new, x="Year", y="Price")

fig.show()
```
